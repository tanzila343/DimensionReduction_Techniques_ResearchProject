```python

######## Apply PCA for 6 data points


# Step-01: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Step-02: Get data
X = np.array([[2, 1],[3,5],[4,3],[5,6],[6,7],[7,8]])

# Scatter plot of the original dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Dataset for PCA')
plt.show()

# Print the dimension before applying dimension reduction techniques

print("Dimensions of datasets without using PCA:", X.shape)


# Step-03: Calculate the mean vector (µ)
mean_vector = np.mean(X, axis=0)
print("\nMean Vector (µ):\n", mean_vector)

# Step-04: Subtract mean vector (µ) from feature vectors
centered_data = X - mean_vector
print("\nCentered Data:\n", centered_data)

# Step-05: Calculate the covariance matrix
covariance_matrix = np.cov(centered_data.T)
print("\nCovariance Matrix:\n", covariance_matrix)

# Step-06: Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

# Step-07: Find the principal component (eigenvector corresponding to the greatest eigenvalue)
principal_component_index = np.argmax(eigenvalues)
principal_component = eigenvectors[:, principal_component_index]
print("\nPrincipal Component:\n", principal_component)

# Step-08: Verify the results: Check if MX = λX holds true for the principal component
verification_result = covariance_matrix @ principal_component - eigenvalues[principal_component_index] * principal_component
print("\nVerification Result (should be close to zero):\n", verification_result)

# Step-09: Project the data onto the principal component
projected_data = centered_data @ principal_component
projected_data = projected_data[:, np.newaxis]  # Ensure it is a 2D array
print("\nProjected Data:\n", projected_data)

# Print the dimension after applying dimension reduction techniques

print("Dimensions of datasets with using PCA:", projected_data.shape)

# Step-10: Plot the results (scatter plot for visualization, if desired)
plt.figure(figsize=(8, 6))
plt.scatter(projected_data[:, 0], np.zeros_like(projected_data[:, 0]), c='blue', marker='o', edgecolors='black')
plt.xlabel('Principal Component 1')
plt.title('Projection onto Principal Component')
plt.show()
```
