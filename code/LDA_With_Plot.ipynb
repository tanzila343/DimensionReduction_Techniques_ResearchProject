**Apply LDA for 20 data points**





```python
######## Apply LDA for 20 data points

import numpy as np
import matplotlib.pyplot as plt

# Step-01: Define the classes
class_1_data = np.array([[200, 120], [180,125], [49, 130], [50, 168], [263, 279], [137, 210],[29,90],[35,78],[16,60],[50,20],[67,10],[47,35],[90,20],[60,15],[30,78],[97,34],[12,80],[46,57],[49,12],[35,67]])
class_2_data = np.array([[100, 120], [120,15], [50, 120], [150, 170], [63, 179], [112, 90],[45,70],[135,210],[160,50],[90,30],[76,25],[67,70],[110,220],[160,153],[130,70],[27,38],[122,180],[146,157],[45,67],[56,10]])

# Step-02: Calculate the class means
mean_class_1 = np.mean(class_1_data, axis=0)
mean_class_2 = np.mean(class_2_data, axis=0)

# Step-03: Calculate the covariance matrix for each class
covariance_class_1 = np.cov(class_1_data.T)
covariance_class_2 = np.cov(class_2_data.T)

# Step-04: Calculate the within-class scatter matrix
within_class_scatter_matrix = covariance_class_1 + covariance_class_2

# Step-05: Calculate the between-class scatter matrix
between_class_scatter_matrix = (len(class_1_data) * (mean_class_1 - mean_class_2).reshape(-1, 1) @
                               (mean_class_1 - mean_class_2).reshape(1, -1))

# Step-06: Calculate eigenvalues and eigenvectors of (SW^(-1) * SB)
eigenvalues, eigenvectors = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix) @ between_class_scatter_matrix)

# Step-08: Obtain the LDA by taking the dot product of eigen vectors and original data
lda_projection_class_1 = class_1_data @ eigenvectors
lda_projection_class_2 = class_2_data @ eigenvectors

# Print the dimensions
print("Original Data Dimension:")
print("Class 1:", class_1_data.shape)
print("Class 2:", class_2_data.shape)
print("\nLDA Projection Dimension:")
print("LDA projection for Class 1:", lda_projection_class_1.shape)
print("LDA projection for Class 2:", lda_projection_class_2.shape)
```
