**SVD For Digit Dataset**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import TruncatedSVD

# Load the digits dataset
digits = load_digits()
X_digits = digits.data
y_digits = digits.target

# Perform SVD for dimensionality reduction
svd_digits = TruncatedSVD(n_components=2)  # Reduce to 2 dimensions
X_svd_digits = svd_digits.fit_transform(X_digits)

# Plotting the results
plt.figure(figsize=(12, 6), facecolor='#2E2E2E')  # Dark background for figure

# Plot original data points
plt.subplot(1, 2, 1, facecolor='#2E2E2E')  # Dark background for subplot
plt.scatter(X_digits[:, 0], X_digits[:, 1], c=y_digits, cmap='tab10')
plt.title('Original Data', color='white')
plt.xlabel('Feature 1', color='white')
plt.ylabel('Feature 2', color='white')
plt.xticks(color='white')
plt.yticks(color='white')

# Plot reduced data points after SVD
plt.subplot(1, 2, 2, facecolor='#2E2E2E')  # Dark background for subplot
plt.scatter(X_svd_digits[:, 0], X_svd_digits[:, 1], c=y_digits, cmap='tab10')
plt.title('SVD Dimensionality Reduction', color='white')
plt.xlabel('SVD Component 1', color='white')
plt.ylabel('SVD Component 2', color='white')
plt.xticks(color='white')
plt.yticks(color='white')

plt.tight_layout()
plt.show()

print("Original data shape:", X_digits.shape)
print("Shape after SVD dimensionality reduction:", X_svd_digits.shape)

```
